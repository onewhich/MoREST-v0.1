```python
def generate_landmarks(representaion_list, cutoff):
    '''
    This function selects landmarks based on the Cartesian distance between representations.
    The choice of cutoff depends on the representation.
    '''
    rep_index = np.array(range(len(representaion_list)))
    i = 0
    if_loop = True
    while if_loop:
        i_rep = representaion_list[rep_index[i]]
        distance = np.array([np.linalg.norm(i_rep - representaion_list[j]) for j in rep_index[i+1:] ])
        rep_index = np.delete(rep_index, np.where(distance < cutoff)[0]+i+1)
        i += 1
        if i >= len(rep_index):
            if_loop = False
        print('i=',i)
        print('index=',len(rep_index))
    return rep_index, representaion_list[rep_index]
```

```python
rep_index, representaion_landmarks = generate_landmarks(representaion_list, cutoff=0.028)
print(len(rep_index))
```
