```python
from ase import units
from scipy import constants
import numpy as np
from copy import deepcopy
from structure_io import read_xyz_file
```

```python
constants.hbar
```

```python
n_beads = 6
C_jk = np.zeros((n_beads,n_beads))
print(C_jk)
for k in range(n_beads):
    if k == 0:
        C_jk[:,k] = np.sqrt(1/n_beads)
    elif k > 0 and k <= (n_beads/2-1):
        for j in range(n_beads):
            C_jk[j,k] = np.sqrt(2/n_beads) * np.cos(2*units.pi*j*k/n_beads)
    elif k == n_beads/2:
        for j in range(n_beads):
            C_jk[j,k] = np.sqrt(1/n_beads)*(-1)**j
    elif k > n_beads/2:
        for j in range(n_beads):
            C_jk[j,k] = np.sqrt(2/n_beads)*np.sin(2*units.pi*j*k/n_beads)
print(C_jk)
```

```python
np.dot(C_jk.T,C_jk)
```

```python
C_jk.T@C_jk
```
