```python
from ase import units
from scipy import constants
kB = units.kB
hbar = units._hbar * units.J
```

```python
def get_n_beads_from_wavenumber(wavenumber, tempertaure):
    '''
                        units
    wavenmuber          cm^-1
    temperature         K
    Hz                  Hz
    constants.c         m/s
    constants.c*100     cm/s
    hbar                eV/Hz
    kB                  eV/K
    '''
    Hz = wavenumber*constants.c*100
    return hbar*Hz/kB/tempertaure
```

```python
get_n_beads_from_wavenumber(793.85,200)
```
